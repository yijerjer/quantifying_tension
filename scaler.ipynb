```python
import numpy as np
import matplotlib.pyplot as plt
from anesthetic.plot import kde_contour_plot_2d
from np_utils import planck_des_data, scale_planck_des_data
```

```python
X0, X0_weights, X1, X1_weights, X_prior, X_prior_weights, _, _, _ = planck_des_data(
    params=["omegabh2", "omegach2", "theta", "tau", "logA", "ns"], div_max=True
)
```

```python

fig, axs = plt.subplots(5, 5, figsize=(15, 15), sharex='col', sharey='row')
for i in range(5):
    for j in range(5):
        if j <= i:
            kde_contour_plot_2d(axs[i, j], X_prior[:, j], X_prior[:, i+1], weights=X_prior_weights)
            kde_contour_plot_2d(axs[i, j], X1[:, j], X1[:, i+1], weights=X1_weights)
            kde_contour_plot_2d(axs[i, j], X0[:, j], X0[:, i+1], weights=X0_weights)
        else:
            fig.delaxes(axs[i, j])
```
