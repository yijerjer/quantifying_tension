```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.random.randn(1000)
```

\int x^2 P(x) dx -> \sum_{bin_i} x_i^2 P(x)

```python
hist, edges = np.histogram(samples, bins=20)
centres = (edges + (edges[1] - edges[0]) / 2)[:-1]
proba = hist / np.sum(hist)

bin_avg_x2 = np.sum(proba * centres**2)
```

\sum_{sample_i} 1/N * x_i^2

```python
sample_avg_x2 = np.sum(samples**2) / samples.size
```

```python
bin_avgs = []
sample_avgs = []
for i in np.linspace(0, 10000, 100):
    samples = np.random.randn(1000)

    hist, edges = np.histogram(samples, bins=20)
    centres = (edges + (edges[1] - edges[0]) / 2)[:-1]
    proba = hist / np.sum(hist)
    bin_avg_x2 = np.sum(proba * centres**2)
    bin_avgs.append(bin_avg_x2)

    sample_avg_x2 = np.sum(samples**2) / samples.size
    sample_avgs.append(sample_avg_x2)

plt.plot(np.linspace(0, 10000, 100), np.array(bin_avgs))
plt.plot(np.linspace(0, 10000, 100), np.array(sample_avgs))
```
